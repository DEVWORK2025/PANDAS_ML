```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A':[1,2,np.nan,40,30,30,20],
                  'B':[5,np.nan,np.nan,20,10,20,10],
                  'C':[1,2,3,12,5,3,4],
                  'D':[10,20,30,120,50,2,1],
                  'E':['a','a','a','b','b','b',np.nan]})
```

```python
for i in df.columns:
    col_na_per = (df[i].isna().sum()/df.shape[0])*100
    if col_na_per > 50:
        df.drop(i, inplace = True)
    elif col_na_per >10:
        if df[i].dtypes=='object':
            df[i].fillna('Unknown', inplace = True)
        else:
            df[i].fillna(df[i].mean(), inplace = True)
```
