```python
def group_days(x):
    if x <10:
        days_group='A'
    elif x<17:
        days_group='B'
    elif x<22:
        days_group = 'C'
    else:
        days_group='D'
        
    return days_group
```

```python
bank_df.day.apply(group_days)
```

```python
bank_df['group_day']=bank_df.day.apply(group_days)
```
