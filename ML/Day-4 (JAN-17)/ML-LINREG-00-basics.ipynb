```python
height=[[4.0],[4.5],[5.0],[5.2],[5.4],[5.8],[6.1],[6.23],[6.4],[6.8]]
weight=[  42 ,  44 , 49, 55  , 53  , 58   , 60  , 64  ,  66 ,  69]

print("height weight")
##list(zip(height, weight))
for row in zip(height, weight):
    print(row[0][0],"->",row[1])
```

```python
# Plotting the height and weight data:

plt.scatter(height, weight, color='c')

plt.xlabel("height")
plt.ylabel("weight")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()

reg.fit(height, weight)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
x1=5.3
reg.coef_[0]*x1  + reg.intercept_
```

```python
m = reg.coef_[0]
c = reg.intercept_
print("slope=",m, "intercept=",c)
```

```python
[reg.coef_[0] * x + reg.intercept_  for x in [ i[0] for i in height]]
```

```python
predicted_values = [reg.coef_ * x[0] + reg.intercept_ for x in [ i[0] for i in height]]
predicted_values
```

```python
plt.scatter(height, weight, color='black')

predicted_values = [reg.coef_[0] * x + reg.intercept_  for x in [ i[0] for i in height]]

plt.plot(height, predicted_values, 'r')

plt.xlabel("height")
plt.ylabel("weight")
```
