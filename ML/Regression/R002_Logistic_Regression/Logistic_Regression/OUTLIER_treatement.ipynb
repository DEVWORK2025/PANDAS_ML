```python
df.Age.mean()
```

```python
### replacing null value of Age column with the mean value of age

df.Age.fillna(df.Age.mean(),inplace  =True)
```

```python
df.isnull().sum()
```

```python
df.Age.plot(kind = 'box')
```

```python
df.Age.median()
```

```python
q1 = df.Age.quantile(0.25)
q1
```

```python
q2 = df.Age.quantile(0.50)
q2
```

```python
q3= df.Age.quantile(0.75)
q3
```

```python
q1, q2,q3
```

```python
## IQR inter quartile range
IQR = q3 - q1
IQR
```

```python
minimum = q1 - (IQR *1.5)
```

```python
maximum = q3 + (IQR *1.5 )
```

```python
df.Age[df.Age<minimum]
```

```python
df.Age[df.Age>maximum]
```

```python
upp=df.Age[df.Age>maximum]
```

```python
out_upper = df.Age[df.Age>maximum]
```

```python
out_lower = df.Age[df.Age < minimum]
```
