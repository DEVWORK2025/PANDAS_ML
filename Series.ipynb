#### change nan values to 0 and other than numeric values to -1

```python
import numpy as np
d1 = {'a':[1,2,'a',4, np.nan],
      'b':[2,np.nan,'v',3,9]}
```

```python
df1 = pd.DataFrame(d1)
```

```python
def CleanData(x):
    if pd.isnull(x):
        return 0
    else:
        if type(x)==str:
            return -1
        else:
            return x 
```

```python
df1["new_a"] = df1['a'].apply(CleanData)
```

```python
for i in df1.columns:
    df1[i] = df1[i].apply(CleanData)
```
